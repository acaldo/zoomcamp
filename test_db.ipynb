```python
# Nombre de las tablas en la base de datos
taxi_data_table = 'green_taxi_data'
zones_table = 'zones'  # Ajusta este nombre según el nombre real de tu tabla de zonas

# Cargar los DataFrames desde las tablas
taxi_data = pd.read_sql_table(taxi_data_table, con=engine)
zones = pd.read_sql_table(zones_table, con=engine)
```

```python
merged_data = pd.merge(taxi_data, zones, left_on='PULocationID', right_on='LocationID', how='left', suffixes=('_taxi', '_pickup'))
merged_data = pd.merge(merged_data, zones, left_on='DOLocationID', right_on='LocationID', how='left', suffixes=('_taxi', '_drop'))
```

```python
merged_data
```

```python
merged_data['PULocationID'] = merged_data['Borough_taxi'] + ' - ' + merged_data['Zone_taxi']
merged_data['DOLocationID'] = merged_data['Borough_drop'] + ' - ' + merged_data['Zone_drop']
```

```python
merged_data
```

```python
# Lista de columnas a eliminar
columnas_a_eliminar = ['Borough_taxi',	'Zone_taxi',	'service_zone_taxi',	'index_drop',	'LocationID_drop',	'Borough_drop',	'Zone_drop',	'service_zone_drop']

# Eliminar las columnas
merged_data = merged_data.drop(columnas_a_eliminar, axis=1)

# Mostrar el resultado
print(merged_data)
```

```python
merged_data=merged_data.drop(['index_taxi',	'LocationID_taxi',	'PU_Location',	'DO_Location'],axis=1)
```
